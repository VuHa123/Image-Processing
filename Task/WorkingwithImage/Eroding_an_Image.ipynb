Ý tưởng: Đây là một phép biến đổi hình thái học, giúp làm mỏng các vùng sáng trong ảnh, loại bỏ các điểm nhiễu nhỏCác điểm nhiễu nhỏ (thường là các điểm sáng) sẽ bị loại bỏ vì giá trị nhỏ nhất trong vùng kernel sẽ là 0 (nếu có nhiễu)., tách rời các đối tượng liên kết với nhau

Cách hoạt động: trượt một phần tử cấu trúc (kernel) qua ảnh(mặc định trong opencv thì stride = 1, anchor point của kerne = (-1, -1): vị trí trung tâm)
- Đặt kernel tại vị trí (0,0) của ảnh.
- Tính giá trị pixel tại điểm neo của kernel dựa trên giá trị nhỏ nhất trong vùng kernel.
- Di chuyển kernel sang vị trí (0,1) và lặp lại quá trình cho đến khi toàn bộ ảnh được xử lý.

```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../image/se1.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img.shape
```

```python
plt.imshow(img)
```

```python
matric = img[75:80, 42:48]
plt.imshow(matric)
```

```python
matric.shape
```

```python
print(matric)
```

```python
kernel = np.ones((3,3), np.uint8)
matric = cv.erode(matric, kernel)
```

```python
print(matric)
```

```python
plt.imshow(matric)
```

```python
kernel = np.ones((5, 5), np.uint8) 
image = cv.erode(img, kernel)  
plt.imshow(image)
```
