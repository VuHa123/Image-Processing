## Đọc ảnh

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../image/OIP.png')
#Tạo một GUI window để hiển thị hình ảnh trên màn hình
cv.imshow('Image', img)
#Để giữ cửa sổ trên màn hình, chúng ta sử dụng waitKey(), khi phát hiện đầu vào đóng, nó sẽ nhả điều khiển:
cv.waitKey(0)#Tham số đầu tiên dùng để giữ màn hình trong ms được chỉ định(>0)
#Để xóa GUI window đã tạo khỏi màn hình
cv.destroyAllWindows()
```

```python
img = cv.imread('../image/OIP.png', cv.IMREAD_COLOR)
#Hiển thị hình ảnh
cv.imshow('Image', img)
cv.waitKey(0)
cv.destroyAllWindows()  
```

```python
#Mở hình ảnh ở chế độ thang độ xám
img = cv.imread('../image/OIP.png', cv.IMREAD_GRAYSCALE)
#Hiển thị hình ảnh
cv.imshow('Image', img)  
cv.waitKey(0)
cv.destroyAllWindows()  
```

```python
img = cv.imread('../image/OIP.png', cv.IMREAD_UNCHANGED)
#Hiển thị hình ảnh
cv.imshow('Image', img)
cv.waitKey(0)
cv.destroyAllWindows()  
```

### sự khác nhau giữa 2 tùy chọn trong hàm cv.imread(): cv.IMREAD_UNCHANGED và cv.IMREAD_COLOR
cv.IMREAD_UNCHANGED giữ nguyên tất cả các kênh của hình ảnh, trong khi cv.IMREAD_COLOR chỉ giữ lại các kênh màu và bỏ qua kênh alpha.

Trong đó, kênh alpha là  một kênh bổ sung trong hình ảnh kỹ thuật số, chứa thông tin về độ trong suốt của từng điểm ảnh. Đây là cách mà các phần mềm đồ họa như Photoshop sử dụng để xác định mức độ trong suốt của các phần khác nhau trong hình ảnh.
